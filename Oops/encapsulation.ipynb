# **Getter and Setter**

```python
class Person:
    def __init__(self,name,age):
        self.__name=name  # This is a Private Variable 
        self.__age=age    # This is a Private Variable 
    
    # Use Getter mathod to get Private variable value
    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    # Use Setter mathod to change Private variable value
    
    def Set_name(self,name):
        self.__name=name
    
    def Set_age(self,age):
        if age >0:
            self.__age=age
    
    
person1=Person("Subhajit",20)
print(person1.get_name())
print(person1.get_age())
person1.Set_name("Pritam")
person1.Set_age(25)

print(person1.get_name())
print(person1.get_age())
    
```

```python
from abc import ABC,abstractmethod
class Vehicle(ABC):
    def driving(self):
        print("The car is ready to use ")
        
    @abstractmethod
    def start_driving(self):
        pass
    
class Car(Vehicle):
    def start_driving(self):
        print("Car Engine started")
        
def operate_vehicle(vehicle):
    vehicle.start_driving()
    vehicle.driving()
    
car=Car()
operate_vehicle(car)
```
