```python
class Rectungal:
    def __init__(self,h,w):
        self.height=h
        self.width=w
    def area(self):
        return self.height*self.width
    
class Circle:
    def __init__(self,r):
        self.radious=r
        
    def area(self):
        return self.radious**2*3.14
    
def Cal_Area(shape):
    return shape.area()

rectungal1= Rectungal(3,4)
circle1=Circle(3)

print(Cal_Area(rectungal1))
print(Cal_Area(circle1))
```

```python
from abc import ABC,abstractmethod

class Veichel(ABC):
    @abstractmethod
    def start_engine(self):
        pass
    
class Car(Veichel):
    def start_engine(self):
        return "Car Engine start"
    
class Motorcycle(Veichel):
    def start_engine(self):
        return "Motorcycle Engine start"
    
def start_veichel(vk):
    print(vk.start_engine())
    
car=Car()
motorcycle=Motorcycle()
start_veichel(car)
start_veichel(motorcycle)
```
