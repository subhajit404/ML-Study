# **Single Inheritan**

```python
class Car:
    def __init__(self,door,windows,engin):
        self.door=door
        self.windows=windows
        self.engin=engin
    def driving(self):
        print(f"This is a {self.engin} car")
class Tesla(Car):
    def __init__(self,door,windows,engin,is_driving):
        super().__init__(door,windows,engin)
        self.is_driving=is_driving
    def selfdriving(self):
        print(f'Tesla support selfdriving : {self.is_driving}')
        
tesla1=Tesla(1,2,"Eletric",True)
tesla1.selfdriving()
tesla1.driving()
```

# **Multiple inherritance**

```python
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        return f"Add me in another inheritance "
    
class Pet:
    def __init__(self,owner):
        self.owner=owner
        
class Dog(Animal,Pet):
    def __init__(self, name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    def speak(self):
        return f"{self.name} says woof to {self.owner}"
    
dog1=Dog("Shero","Subha")
print(dog1.speak())
```
