```python
# leap Year
year = int(input("enter the year "))

if year %4==0:
    if year %100==0:
        if year%400==0:
            print(year,"This is a leap year")
        else :
            print(year,"This is not a leap year")
    else:
        print(year,"this is leap year")
```
