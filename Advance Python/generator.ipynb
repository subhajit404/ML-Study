```python
def square():
    for i in range(9):
        yield i**2
        
```

```python
for i in square():
    print(i)
```

```python
def read_large_file(file_name):
    with open(file_name,'r') as file:
        for line in file:
            yield line.strip()
```

```python
file="example.txt"
for line in read_large_file(file):
    print(line)
```
