```python
## Decorator

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
def outer(func):
    def inner():
        print("Hello")
        func()
        print("World !")
    return inner
```

```python
@outer
def say_helloo():
    print("Hello dunia!")
```

```python

say_helloo()
```
