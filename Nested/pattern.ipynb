```python
## stars
for i in range (1,5):
    for j in range(1,5):
        print("*",end=" ")
    print()
```

```python
## number pattern
for i in range(1,5):
    for j in range(1,5):
        print(j*10, end=" ")
    print("")
```

```python
for i in range(1,5):
    for j in range(1,i+1):
        print("*", end=" ")
    print("")
```
