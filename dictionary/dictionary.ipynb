```python
student = {
    "student1": {"Name": "John Doe", "Age": 20, "Major": "Computer Science"},
    "student2": {"Name": "Jane Smith", "Age": 22, "Major": "Mathematics"},
    "student3": {"Name": "Alice Johnson", "Age": 21, "Major": "Physics"}
}

for student_id, details in student.items():
    print(f"Student ID: {student_id}")
    for key, value in details.items():
        print(f"{key}: {value}")
    print()  # Print a 
```
