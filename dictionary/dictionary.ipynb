```python
student = {
    "student1": {"Name": "John Doe", "Age": 20, "Major": "Computer Science"},
    "student2": {"Name": "Jane Smith", "Age": 22, "Major": "Mathematics"},
    "student3": {"Name": "Alice Johnson", "Age": 21, "Major": "Physics"}
}

for student_id, details in student.items():
    print(f"Student ID: {student_id}")
    for key, value in details.items():
        print(f"{key}: {value}")
    print()  # Print a 
```

```python
# for using zip
# using zip
days = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]
z= {i:j for (i,j) in zip(days,temp_C)}
print(z)
```

```python
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

z= {i:j for (i,j) in products.items() if j>0}
print(z)
```
